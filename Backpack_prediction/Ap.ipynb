```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_train.columns
```

```python
df_train['Weight Capacity (kg)'].fillna(df_train['Weight Capacity (kg)'].mean(),inplace=True)
df_test['Weight Capacity (kg)'].fillna(df_test['Weight Capacity (kg)'].mean(),inplace=True)
df_train['Brand'].fillna(df_train['Brand'].mode()[0],inplace=True)
df_test['Brand'].fillna(df_test['Brand'].mode()[0],inplace=True)
df_train['Material'].fillna(df_train['Material'].mode()[0],inplace=True)
df_test['Material'].fillna(df_test['Material'].mode()[0],inplace=True)
df_train['Size'].fillna(df_train['Size'].mode()[0],inplace=True)
df_test['Size'].fillna(df_test['Size'].mode()[0],inplace=True)
df_train['Laptop Compartment'].fillna(df_train['Laptop Compartment'].mode()[0],inplace=True)
df_test['Laptop Compartment'].fillna(df_test['Laptop Compartment'].mode()[0],inplace=True)
df_train['Waterproof'].fillna(df_train['Waterproof'].mode()[0],inplace=True)
df_test['Waterproof'].fillna(df_test['Waterproof'].mode()[0],inplace=True)
df_train['Style'].fillna(df_train['Style'].mode()[0],inplace=True)
df_test['Style'].fillna(df_test['Style'].mode()[0],inplace=True)
df_train['Color'].fillna(df_train['Color'].mode()[0],inplace=True)
df_test['Color'].fillna(df_test['Color'].mode()[0],inplace=True)
```
